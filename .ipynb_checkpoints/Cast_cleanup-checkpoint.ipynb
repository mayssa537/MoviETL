## Cleanup cast csv to extract actor names and reduce size

```python
import pandas as pd
import ast
```

```python
cast_csv = "../cast.csv"


cast_df = pd.read_csv(cast_csv)

cast_df = cast_df[['cast', 'id']]
cast_df.head()
```

```python
cast_list = list(cast_df['cast'])

cast_split = []
cast_replace = []
cast_final = []

for item in cast_list:
    cast_split.append(str(item).split('}'))
    
# cast_    
for item_list in cast_split:    
    cast_replace = []
    
    for item in item_list: 
        new_item=str(item)
        new_item=new_item.replace('[','')
        new_item=new_item.replace('\\','')
        new_item=new_item.replace(', {' ,'{')
        new_item=new_item.replace(']}' ,'')
        new_item=new_item.replace('} ' ,'}')
        new_item=new_item.replace('"' ,'')
        new_item=new_item.replace("'" ,"")       
        new_item=new_item.replace("cast_id: " ,"'cast_id': '")    
        new_item=new_item.replace(", character: " ,"', 'character': '")                
        new_item=new_item.replace(", credit_id: " ,"', 'credit_id': '")                
        new_item=new_item.replace(", gender: " ,"', 'gender': '")    
        new_item=new_item.replace(", id: " ,"', 'id': '")    
        new_item=new_item.replace(", name: " ,"', 'name': '")    
        new_item=new_item.replace(", order: " ,"', 'order': '")               
        new_item=new_item.replace(", profile_path: " ,"', 'profile_path': '") 
        new_item=new_item.replace(']','')
        
        if new_item == ']}':
            new_item = '{}'
            
        if new_item != '{}' and new_item != '':
            new_item=new_item + "'}"
            
            try:
                cast_replace.append(ast.literal_eval(new_item))
                
            except:
                cast_replace.append({'name': ''})
        
    cast_final.append(cast_replace)
```

```python
cast_final[0]
```

```python
cast_column = []
cast_name = []

for item_list in cast_final:
    cast_name = []
    for item in item_list:
        for k, v in item.items():
            if k == 'name':
                cast_name.append(v)
    cast_column.append(cast_name)
```

```python
id_list = list(cast_df['id'])
```

```python
cast_final = []
index_final = []

for index, value in enumerate(cast_column):
    
    for item in value:
        cast_final.append(item)
        index_final.append(id_list[index])
```

```python
print(len(cast_final), len(index_final))
```

```python
final_cast_df = pd.DataFrame({'name':cast_final, 'movie_id':index_final})
```

```python
final_cast_df.head(30)
```

```python
final_cast_df = final_cast_df[final_cast_df['name'] != '']
```

```python
final_cast_df.to_csv("Resources/cast_new.csv", index=False)
```

## Extract genre table from movie data

```python
movie_csv = "Resources/movies_metadata.csv"

movie_df = pd.read_csv(movie_csv)
```

```python
genre_list = list(movie_df['genres'])
```

```python
genre_list
```

```python
genre_dict = []

for item in genre_list:
    genre_dict.append(ast.literal_eval(item))
```

```python
genre_dict
```

```python
genre_name = []
genre_column = []

for item_list in genre_dict:
    genre_name = []
    for item in item_list:
        for k, v in item.items():
            if k == 'name':
                genre_name.append(v)
    genre_column.append(genre_name)
```

```python
genre_column
```

```python
genre_id_list = list(movie_df['id'])

genre_final = []
genre_id_final = []

for index, value in enumerate(genre_column):
    for item in value:
        genre_final.append(item)
        genre_id_final.append(genre_id_list[index])
```

```python
final_genre_df = pd.DataFrame({'name':genre_final, 'movie_id':genre_id_final})

final_genre_df = final_genre_df[final_genre_df['name'] != '']
```

```python
len(final_genre_df)

final_genre_df.head()
```

```python
final_genre_df.to_csv("Resources/genre_new.csv", index=False)
```
