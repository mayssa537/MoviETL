## Cleanup cast csv to extract actor names and reduce size

```python
import pandas as pd
import ast
```

```python
cast_csv = "../credits.csv"

cast_df = pd.read_csv(cast_csv, encoding = 'latin-1')

cast_df = cast_df[['cast', 'id']]
cast_df.head()
```

```python
cast_list = list(cast_df['cast'])

cast_list_list = []

for item in cast_list:
    try:
        cast_list_list.append(ast.literal_eval(item))
    except:
        cast_list_list.append('')
```

```python
# type(cast_list_list[0])
cast_list_list[0]
```

```python
cast_column = []
cast_name = []

for item_list in cast_list_list:
    cast_name = []
    try:
        for item in item_list:
            for k, v in item.items():
                if k == 'name':
                    cast_name.append(v)
    except:
        cast_name.append('')
                    
    cast_column.append(cast_name)
```

```python
cast_column[0]
```

```python
id_list = list(cast_df['id'])

cast_final = []
index_final = []

for index, value in enumerate(cast_column):
    
    for item in value:
        cast_final.append(item)
        index_final.append(id_list[index])
```

```python
final_cast_df = pd.DataFrame({'name':cast_final, 'movie_id':index_final})
final_cast_df = final_cast_df[final_cast_df['name'] != '']
```

```python
final_cast_df.head(30)
```

```python
final_cast_df.to_csv("Resources/cast_new.csv", index=False)
```

## Extract genre table from movie data

```python
movie_csv = "../movies_metadata.csv"

movie_df = pd.read_csv(movie_csv, encoding = 'latin-1')
```

```python
genre_list = list(movie_df['genres'])
```

```python
genre_list
```

```python
genre_dict = []

for item in genre_list:
    genre_dict.append(ast.literal_eval(item))
```

```python
genre_dict[0]
```

```python
genre_name = []
genre_column = []

for item_list in genre_dict:
    genre_name = []
    for item in item_list:
        for k, v in item.items():
            if k == 'name':
                genre_name.append(v)
    genre_column.append(genre_name)
```

```python
genre_column[0]
```

```python
genre_id_list = list(movie_df['id'])

genre_final = []
genre_id_final = []

for index, value in enumerate(genre_column):
    for item in value:
        genre_final.append(item)
        genre_id_final.append(genre_id_list[index])
```

```python
final_genre_df = pd.DataFrame({'name':genre_final, 'movie_id':genre_id_final})

final_genre_df = final_genre_df[final_genre_df['name'] != '']
```

```python
len(final_genre_df)

final_genre_df.head()
```

```python
final_genre_df.to_csv("Resources/genre_new.csv", index=False)
```

## Extract production table from movie data

```python
production_list = list(movie_df['production_companies'])
```

```python
production_list
```

```python
production_dict = []

for item in production_list:
    try:
        production_dict.append(ast.literal_eval(item))
    except:
        production_dict.append('')
```

```python
production_dict[0]
```

```python
production_name = []
production_column = []

for item_list in production_dict:
    production_name = []
    try:
        for item in item_list:
            for k, v in item.items():
                if k == 'name':
                    production_name.append(v)
    except:
        production_name.append('')
        
    production_column.append(production_name)        
```

```python
production_column[0]
```

```python
production_id_list = list(movie_df['id'])

production_final = []
production_id_final = []

for index, value in enumerate(production_column):
    
    for item in value:
        production_final.append(item)
        production_id_final.append(production_id_list[index])
```

```python
final_production_df = pd.DataFrame({'name': production_final, 'movie_id': production_id_final})

final_production_df = final_production_df[final_production_df['name'] != '']
```

```python
final_production_df.head()
```

```python
final_production_df.to_csv("Resources/production_new.csv", index=False)
```

## Extract average rating table from rating csv

```python
ratings_csv = "../ratings.csv"

ratings_df = pd.read_csv(ratings_csv, encoding = 'latin-1')

ratings_df = ratings_df[['rating', 'movieId']]

ratings_df.columns = ['rating', 'movie_id']
```

```python
ratings_df.head()
```

```python
final_ratings_df = ratings_df.groupby(['movie_id']).mean()

final_ratings_df = final_ratings_df.dropna(how='any')

final_ratings_df.head()
```

```python
final_ratings_df.to_csv("Resources/ratings_new.csv", index=True)
```
